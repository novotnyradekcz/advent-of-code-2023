```python
def to_binary_string_list(number, length, memo={}):
    if (number, length) in memo and number != 0:
        return (memo[(number, length)].copy())
    result = bin(number)[2:]
    # result = '0' * (length - len(result)) + result
    list_result = ['0'] * (length - len(result))
    for digit in result:
        list_result.append(digit)
    memo[(number, length)] = list_result
    # print(memo[(number, length)])
    return list_result + ['0']

def test_string_groups(string_number, groups, memo={}):
    if (string_number, tuple(groups)) in memo:
        return memo[(string_number, tuple(groups))]

    def recursive_test(binary_string, groups):
        i = 0
        while binary_string[i] == '0':
            i += 1
            if i == len(binary_string):
                return False
        j = 0
        while binary_string[i + j] == '1':
            j += 1
            if i + j == len(binary_string):
                break
        if j == groups[0]:
            if len(groups) == 1:
                memo[(binary_string, tuple(groups))] = binary_string[i + j:].count('1') == 0
                return memo[(binary_string, tuple(groups))]
            elif i + j == len(binary_string):
                return False
            else:
                result = recursive_test(binary_string[i + j:], groups[1:])
                memo[(binary_string, tuple(groups))] = result
                return result
        else:
            memo[(binary_string, tuple(groups))] = False
            return False

    if string_number.count('1') == 0:
        result = not groups
        memo[(string_number, tuple(groups))] = result
        return result

    result = recursive_test(string_number, groups)
    memo[(string_number, tuple(groups))] = result
    return result

def part_one(input_lines):
    spring_map = []
    broken_groups = []
    for line in input_lines:
        split_line = line.strip('\n').split()
        binary_string = split_line[0].replace('#', '1').replace('.', '0')
        spring_map.append(binary_string + '?' + binary_string)
        groups = [int(group) for group in split_line[1].split(',')]
        broken_groups.append(groups * 2)

    all_the_ways = []
    total = 0
    memo = {}
    binary_memo = {}
    for i in range(len(spring_map)):
        print(i, end="\r")
        ways = 0
        unknowns = spring_map[i].count('?')
        
        for j in range(2 ** unknowns):
            current_tested_values = to_binary_string_list(j, unknowns, binary_memo)
            arrangement = ''.join(
                spring_map[i][k] if spring_map[i][k] != '?' else current_tested_values.pop(0)
                for k in range(len(spring_map[i]))
            )
            if test_string_groups(arrangement, broken_groups[i], memo):
                ways += 1
        all_the_ways.append(ways)
        total += ways
    print(all_the_ways)
    return total

print("Example: " + str(part_one(linesEx)))
print("Real: " + str(part_one(lines)))
```
