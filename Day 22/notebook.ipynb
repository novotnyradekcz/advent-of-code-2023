```python
def parseInput(inputLines):
	bricks = []
	for line in inputLines:
		splitLine = line.split('~')
		brick = []
		for part in splitLine:
			stringPart = part.split(',')
			brickPart = (int(stringPart[0]), int(stringPart[1]), int(stringPart[2]))
			brick.append(brickPart)
		if brick[0] == brick[1]:
			brick.pop()
		elif brick[0][0] != brick[1][0]:
			if brick[0][0] < brick[1][0]:
				for i in range(brick[0][0] + 1, brick[1][0]):
					brick.append((i, brick[0][1], brick[0][2]))
			else:
				for i in range(brick[1][0] + 1, brick[0][0]):
					brick.append((i, brick[0][1], brick[0][2]))
		elif brick[0][1] != brick[1][1]:
			if brick[0][1] < brick[1][1]:
				for i in range(brick[0][1] + 1, brick[1][1]):
					brick.append((brick[0][0], i, brick[0][2]))
			else:
				for i in range(brick[1][1] + 1, brick[0][1]):
					brick.append((brick[0][0], i, brick[0][2]))
		elif brick[0][2] != brick[1][2]:
			if brick[0][2] < brick[1][2]:
				for i in range(brick[0][2] + 1, brick[1][2]):
					brick.append((brick[0][0], brick[0][1], i))
			else:
				for i in range(brick[1][2] + 1, brick[0][2]):
					brick.append((brick[0][0], brick[0][1], i))
		bricks.append(brick)
	return(bricks)

print(parseInput(linesEx))

def fall(bricks):
	for brick in bricks:
		brick.sort(key=lambda x: x[2])
	bricks.sort(key=lambda x: x[0][2])

	for i in range(len(bricks)):
		for j in range(i):
			
```
